**Построение гистограмм числовых переменных**

```python
# Строим гистограмму для столбца 'age' в DataFrame df
plt.hist(df['age'], bins=20, color='blue', edgecolor='darkblue')
plt.title('Age')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(8,6))
df['gender'].value_counts().plot(kind='bar', color='blue', edgecolor='darkblue')
plt.title('Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(8,6))
df['hypertension'].value_counts().plot(kind='bar', color='blue', edgecolor='darkblue') #гипертония
plt.title('Hypertension')
plt.xlabel('Hypertension')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(8,6))
df['smoking_history'].value_counts().plot(kind='bar', color='blue', edgecolor='darkblue')
plt.title('Smoking History')
plt.xlabel('Smoking History')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(8,6))
df['diabetes'].value_counts().plot(kind='bar', color='blue', edgecolor='darkblue')
plt.title('Diabetes')
plt.xlabel('Diabetes')
plt.ylabel('Count')
plt.show()
```

```python
diabetes_values = [df[df['diabetes'] == 0]['age'], df[df['diabetes'] == 1]['age']]# Разделим данные по столбцу 'diabetes'
plt.figure(figsize=(8, 6))
plt.boxplot(diabetes_values, labels=['No Diabetes', 'Diabetes'])
plt.title('Age and Diabetes')
plt.xlabel('Diabetes')
plt.ylabel('Age')
plt.show()
```

```python
bmi_values = [df[df['diabetes'] == 0]['bmi'], df[df['diabetes'] == 1]['bmi']]
plt.figure(figsize=(8, 6))
plt.boxplot(bmi_values, labels=['No Diabetes', 'Diabetes'])
plt.title('BMI and Diabetes')
plt.xlabel('Diabetes')
plt.ylabel('BMI')
plt.show()
```

```python
HbA1c_values = [df[df['diabetes'] == 0]['HbA1c_level'], df[df['diabetes'] == 1]['HbA1c_level']]
plt.figure(figsize=(8, 6))
plt.boxplot(HbA1c_values, labels=['No Diabetes', 'Diabetes'])
plt.title('HbA1c Level and Diabetes')
plt.xlabel('Diabetes')
plt.ylabel('HbA1c Level')
plt.show()
```

```python
blood_glucose_values = [df[df['diabetes'] == 0]['blood_glucose_level'], df[df['diabetes'] == 1]['blood_glucose_level']]
plt.figure(figsize=(8, 6))
plt.boxplot(blood_glucose_values, labels=['No Diabetes', 'Diabetes'])
plt.title('Blood Glucose Level and Diabetes')
plt.xlabel('Diabetes')
plt.ylabel('Blood Glucose Level')
plt.show()
```

```python
gender_diabetes_counts = pd.crosstab(df['gender'], df['diabetes'])
gender_diabetes_counts.plot(kind='bar', stacked=False, figsize=(8, 6), color=['blue', 'orange'])
plt.title('Gender and Diabetes')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

```python
hypertension_diabetes_counts = pd.crosstab(df['hypertension'], df['diabetes'])
hypertension_diabetes_counts.plot(kind='bar', stacked=False, figsize=(8, 6), color=['blue', 'orange'])
plt.title('Hypertension and Diabetes')
plt.xlabel('Hypertension')
plt.ylabel('Count')
plt.show()
```

```python
heart_disease_diabetes_counts = pd.crosstab(df['heart_disease'], df['diabetes']) #сердечно-сосудистые заболевания
heart_disease_diabetes_counts.plot(kind='bar', stacked=False, figsize=(8, 6), color=['blue', 'orange'])
plt.title('Heart Disease and Diabetes')
plt.xlabel('Heart Disease')
plt.ylabel('Count')
plt.show()
```

```python
smoking_history_diabetes_counts = pd.crosstab(df['smoking_history'], df['diabetes'])
smoking_history_diabetes_counts.plot(kind='bar', stacked=False, figsize=(8, 6), color=['blue', 'orange'])
plt.title('Smoking History and Diabetes')
plt.xlabel('Smoking History')
plt.ylabel('Count')
plt.show()
```

```python
import phik
from phik.report import plot_correlation_matrix
from phik import report

phik_overview = df.phik_matrix()

plot_correlation_matrix(phik_overview.values, 
                        x_labels=phik_overview.columns, 
                        y_labels=phik_overview.index, 
                        vmin=0, vmax=1, color_map="magma_r", 
                        fontsize_factor=0.8, figsize=(10, 10))
plt.tight_layout()
```
